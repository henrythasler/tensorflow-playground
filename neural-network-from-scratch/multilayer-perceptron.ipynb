```python
# TensorFlow and tf.keras
import tensorflow as tf

import numpy as np 
import numpy.matlib as matlib
import matplotlib.pyplot as plt
```

```python
dataset = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = dataset.load_data()
train_images = train_images / 256.0
test_images = test_images / 256.0
```

```python
def plot_flat_images(images, dim, cmap=plt.cm.viridis):
    plt.figure(figsize=(10,10))
    items=16
    for i in range(items):
        plt.subplot(4,4,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(images[i].reshape(dim), cmap)
    plt.show()
    
def plot_vector(vector, cmap=plt.cm.viridis):
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(vector.reshape(vector.shape[0], 1), cmap)
    plt.colorbar()
```
